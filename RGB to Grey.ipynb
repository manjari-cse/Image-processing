```
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from skimage import color
from skimage import io
im1 = io.imread('/content/klcc_a.png')
im2 = io.imread('/content/klcc_b.png')
im1=color.rgb2gray(im1)
im1= im1*255
im1= im1.astype(np.uint8)


im2=color.rgb2gray(im2)
im2= im2*255
im2= im2.astype(np.uint8)

im3 = np.bitwise_xor(im2,im1)

plt.imshow(im1)
plt.show()
plt.imshow(im2)
plt.show()
plt.imshow(im3,cmap='gray')
plt.show()
```
