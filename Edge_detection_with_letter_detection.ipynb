### **Different edge detection methods**

```
import cv2
from skimage import data
from matplotlib import pyplot as plt
from skimage.io import imread
from scipy.ndimage.filters import convolve
from skimage.filters import sobel
from skimage.feature import canny
import numpy as np
I = imread('/content/letter K.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
plt.imshow(I,cmap='gray')
plt.show()
k = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]]) # Defining filter
I1 = convolve(I, k, mode='constant', cval=0.0)
plt.imshow(I1,cmap='gray')
plt.show()
I2 = sobel(I)
plt.imshow(I2,cmap='gray')
plt.show()
I3 = cv2.Canny(I,100,200)
plt.imshow(I3,cmap='gray')
plt.show()
```

```
import cv2
from skimage.feature import canny
from skimage.filters import sobel
import pytesseract
from PIL import Image
from matplotlib import pyplot as plt
I = imread('/content/02_19.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
plt.imshow(I,cmap='gray')
plt.show()

I3 = cv2.Canny(I,100,200)
plt.imshow(I3,cmap='gray')
plt.show()




text = pytesseract.image_to_string(I3)
print(text)

```

```
from skimage import data
from skimage.filters import threshold_otsu
from skimage.transform import rescale, resize, downscale_local_mean
Image = imread('/content/02_01.jpg')
plt.imshow(Image)
plt.show()
Image = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
image_rescaled =  downscale_local_mean(Image, (1, 1))
plt.imshow(image_rescaled)
plt.show()
#Image = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
thres = threshold_otsu(image_rescaled)
Thresholded_Image = image_rescaled>=thres
plt.imshow(Thresholded_Image, cmap='gray')
plt.show()

text = pytesseract.image_to_string(Thresholded_Image)
print(text)
```
